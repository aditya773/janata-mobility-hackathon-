```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("train_Wc8LBpr.csv")
```

```python
df
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df
```

```python
df.corr()
```

```python
del df['Customer_Since_Months']
del df['Var2']
del df['Var3']
del df['Var1']
del df['Life_Style_Index']
```

```python
df
```

```python
del df['Gender']
del df['Trip_ID']
```

```python
df
```

```python
df['Type_of_Cab'] = df['Type_of_Cab'].map({'A':'1','B':'2','C':'3','D':'4','E':'5'})
```

```python
df['Confidence_Life_Style_Index'] = df['Confidence_Life_Style_Index'].map({'A':'a','B':'b','C':'c'})
```

```python
df['Destination_Type'].value_counts()
```

```python
df['Destination_Type'].replace('A','Dest1',inplace=True)
df['Destination_Type'].replace('B','Dest1',inplace=True)
df['Destination_Type'].replace('C','Dest1',inplace=True)
df['Destination_Type'].replace('D','Dest1',inplace=True)
df['Destination_Type'].replace('E','Dest1',inplace=True)
df['Destination_Type'].replace('F','Dest2',inplace=True)
df['Destination_Type'].replace('G','Dest2',inplace=True)
df['Destination_Type'].replace('H','Dest2',inplace=True)
df['Destination_Type'].replace('I','Dest2',inplace=True)
df['Destination_Type'].replace('J','Dest2',inplace=True)
df['Destination_Type'].replace('K','Dest3',inplace=True)
df['Destination_Type'].replace('L','Dest3',inplace=True)
df['Destination_Type'].replace('M','Dest3',inplace=True)
df['Destination_Type'].replace('N','Dest3',inplace=True)
```

```python
df
```

```python
df['Destination_Type'].value_counts()
```

```python
df_cab = pd.get_dummies(df['Type_of_Cab'])
df = df.join(df_cab)
del df['Type_of_Cab']
```

```python
df_con = pd.get_dummies(df['Confidence_Life_Style_Index'])
df = df.join(df_con)
del df['Confidence_Life_Style_Index']
```

```python
df_des = pd.get_dummies(df['Destination_Type'])
df = df.join(df_des)
del df['Destination_Type']
```

```python
df
```

```python
x = df.drop('Surge_Pricing_Type',axis=1)
y = df['Surge_Pricing_Type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(max_depth=10,n_estimators=300,criterion='entropy')
```

```python
rfc.fit(x_train,y_train)
```

```python
predictions = rfc.predict(x_test)
```

```python
y_predict = rfc.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(confusion_matrix(predictions,y_test))
```

```python
print(classification_report(predictions,y_test))
```

```python
print("training accuracy:{}".format(accuracy_score(y_predict,y_train)))
print("test accuracy:{}".format(accuracy_score(predictions,y_test)))
```
